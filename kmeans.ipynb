```python
from pcl.kmeans import KMeans
import torch
import matplotlib.pyplot as plt
```

```python
km = KMeans(4, 10)
x = torch.cat([torch.randn(30, 2)/ 3, torch.randn(100, 2) + 1, + torch.randn(40, 2) / 10 + 2]) + 10
cl1, centroids1, n_points1 = km(x)
y = torch.randn(100, 2) + 5
cl2, centroids2, n_points2 = km(y)
x = torch.cat([x, y])
cl = torch.cat([cl1, cl2])
centroids = torch.cat([centroids1, centroids2])
n_points = torch.cat([n_points1, n_points2])
plt.scatter(x[:, 0], x[:, 1], c=cl)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=300)
```
